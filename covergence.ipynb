```python
import non_parametric
import numpy as np
import matplotlib.pyplot as plt
import tqdm
```

```python
size = 4000
l = np.zeros(size)
h = 0.3
norm = np.random.normal(size = size)
for i in tqdm.tqdm(range(size)):
    est = non_parametric.Estimator(norm[:i], h, 100)
    l[i] = est.Error(1/np.sqrt(2*np.pi))**2
```

```python
order_log = [np.log(1/(i*2*h) + h**2) for i in range(1,size+1)]
l_log = [np.log(i) for i in l ]
order = [1/(i*2*h) + h**2 for i in range(1,size+1)]
proportion = np.log(l/order)
inv_proportion = np.log(order/l)
```

```python
plt.plot(range(size), proportion, color = "m", lw = 1)
```

```python
plt.plot(range(size), l_log, color = "m", lw = 1)
plt.plot(range(size), order, color = "c", lw = 1)
plt.savefig("covergence_log.png", format = "png", dpi = 1000)
```
