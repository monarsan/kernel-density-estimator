```python
import non_parametric
import numpy as np
import matplotlib.pyplot as plt
```

```python
size = 1000
norm = np.random.normal(size=size)
est = non_parametric.Estimator(norm, 0.3, 100)
```

```python
binorm =np.concatenate([np.random.normal(3, 0.3, size=int(size*0.4)), np.random.normal(-3, 1, size=int(size*0.6))], axis = 0)
est_bi = non_parametric.Estimator(binorm, 0.3, 100)
```

```python
est_bi.rosenblattPlot()
plt.hist(binorm, bins = 100, color = "c", density=True, label = "true")
plt.legend()
plt.savefig("h03/rosenblatt-bi.png", format="png", dpi=500)
```
